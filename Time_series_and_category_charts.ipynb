# **Visualization**

**Sales by Product Category**

```python
category_sales= df.groupby("Product_Category")["Sales_Amount"].sum().sort_values(ascending=False)
print("\033[1mSales by Product Category:\033[0m")
print(category_sales)
```

```python
plt.figure(figsize=(6, 6))
category_sales.plot(kind="bar", color="c", alpha=0.5, label="Sales Amount")

plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales (Taka)")
plt.legend(title="Metric")
plt.tight_layout()
plt.savefig('category_sales_bar.png')
plt.show()
```

**Chart Discussion – Sales by Product Category**

- **Chart type:** Vertical bar chart — ideal for comparing named categories side by side.
- **X-axis:** Product Category names — categorical.
- **Y-axis:** Total Sales in Taka — quantitative, clearly labelled.
- **Legend:** Added via `label=` and `plt.legend()` to identify the metric being plotted.
- **Layout:** `tight_layout()` prevents label clipping along edges.

 **Best Performer**

```python
top_sales_rep = df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values(ascending=False)
print(top_sales_rep)
```

```python
top_sales_rep.plot(kind="bar", color="m", alpha=0.5, label="Sales Amount")
plt.title("Best Performer by Sales Rep")
plt.xlabel("Sales Representative")
plt.ylabel("Total Sales (Taka)")
plt.xticks(rotation=30, ha="right")
plt.legend(title="Metric")
plt.tight_layout()
plt.savefig('best_performer_bar.png')
plt.show()
```

**Chart Discussion – Best Performer**

- **Chart type:** Vertical bar chart — best for ranking individuals side by side.
- **X-axis:** Sales Rep names — categorical, rotated 30° so long names don't overlap.
- **Y-axis:** Total Sales in Taka — quantitative.
- **Legend:** Labels the metric for clarity.
- **Layout:** `tight_layout()` + `ha='right'` alignment keeps tick labels neat.

**Sales Over Time (Monthly Line Chart)**

```python

df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])
monthly_sales = df.resample('ME', on='Sale_Date')['Sales_Amount'].sum()

plt.figure(figsize=(8, 4))
monthly_sales.plot(kind='line', marker='o', color='steelblue', label='Monthly Sales')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales (Taka)')
plt.xticks(rotation=45)
plt.legend(title='Period')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('monthly_sales_line.png')
plt.show()
print('Chart saved as monthly_sales_line.png')
```

**Chart Discussion – Monthly Sales Over Time (Line Chart)**

- **Chart type:** Line chart — the natural choice for time-series data; connected points reveal trends and seasonality across months.
- **Markers:** Circle markers (`marker='o'`) make individual monthly data points easy to identify.
- **X-axis:** Month dates, rotated 45° to prevent label overlap.
- **Y-axis:** Total Sales in Taka — quantitative.
- **Grid:** Horizontal dashed grid lines (`linestyle='--'`) help the eye trace values across the chart.
- **Legend:** Identifies the line as 'Monthly Sales' so the chart is self-explanatory.

**Quarterly Sales Aggregation**

```python

quarterly_sales = df.resample('QE', on='Sale_Date')['Sales_Amount'].sum()

plt.figure(figsize=(8, 5))
quarterly_sales.plot(kind='bar', color='coral', alpha=0.7, label='Quarterly Sales')
plt.title('Quarterly Sales Aggregation')
plt.xlabel('Quarter')
plt.ylabel('Total Sales (Taka)')
plt.xticks(rotation=45)
plt.legend(title='Period')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('quarterly_sales_bar.png')
plt.show()
print('Chart saved as quarterly_sales_bar.png')
```

**Chart Discussion – Quarterly Sales Aggregation**

- **Chart type:** Bar chart — aggregates monthly values into 4 quarters for a higher-level business view.
- **X-axis:** Quarter labels (e.g., 2023 Q1), rotated 45° for readability.
- **Y-axis:** Total Sales in Taka.
- **Grid:** Horizontal dashed grid lines for easier cross-bar value comparison.
- **Legend:** Labels the bars as 'Quarterly Sales'.

**Payment Method**

```python
payment_sales = df["Payment_Method"].value_counts()
payment_sales.plot(
    kind="pie",
    autopct="%1.1f%%",
    explode=(0.1, 0, 0),
    startangle=140,
    legend=True
)
plt.title("Payment Method Share")
plt.xlabel("")
plt.ylabel("")
plt.legend(title="Payment Method", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.savefig('payment_method_pie.png', bbox_inches='tight')
plt.show()
```

**Chart Discussion – Payment Method Share**

- **Chart type:** Pie chart — shows each method's proportional share of the whole (part-to-whole relationship).
- **Labels:** `autopct='%1.1f%%'` prints the exact percentage inside each slice.
- **Explode:** The first slice is pulled out slightly to highlight the most-used payment method.
- **Legend:** Placed outside the chart via `bbox_to_anchor` so it doesn't overlap any slice.
- **Start angle:** `startangle=140` rotates the chart so the largest slice begins near the top.

**Export Short Summary**

```python
summary = f"""
Sales Summary
=============
Total Sales Amount   : {df['Sales_Amount'].sum():.2f}
Top Region           : {df.groupby('Region')['Sales_Amount'].sum().idxmax()}
Top Product Category : {df.groupby('Product_Category')['Sales_Amount'].sum().idxmax()}
Top Sales Rep        : {df.groupby('Sales_Rep')['Sales_Amount'].sum().idxmax()}
Date Range           : {df['Sale_Date'].min().date()} to {df['Sale_Date'].max().date()}
"""
print(summary)

with open('sales_summary.txt', 'w') as f:
    f.write(summary)
print('Summary saved as sales_summary.txt')
```

**Chart Choice & Formatting Discussion**

- **Line Chart** is used for sales over time (monthly/quarterly) because it clearly shows trends and changes across a continuous time axis.
- **Bar Chart** is used for category comparisons (Product Category, Sales Rep, Quarterly Sales) because it makes it easy to compare discrete groups side by side.
- **Pie Chart** is used for Payment Method share because it shows how the whole is divided among a few categories.
- All charts include a **title**, clear **axis labels**, and appropriate **tick formatting** (e.g., rotated dates) to ensure readability.
- Charts are **saved as PNG files** for easy sharing and inclusion in reports.
